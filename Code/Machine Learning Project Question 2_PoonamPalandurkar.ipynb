# 2.4) Plot the word cloud of each of the three speeches. (after removing the stopwords)

```python
#!pip install wordcloud
```

```python
df = pd.read_excel("Project_Speech.xlsx")
df.head()
```

```python
stopwords = nltk.corpus.stopwords.words('english') +list(string.punctuation)+['--', '``', "''"]
```

```python
#Removing stop words (extended list as above) from the corpus 

corpus = df['Speech'].apply(lambda x: ' '.join([z for z in x.split() if z not in stopwords])) 
corpus
```

```python
wc_a = ' '.join(corpus)
```

```python
# Word Cloud 
from wordcloud import WordCloud
wordcloud = WordCloud(width = 3000, height = 3000, 
                background_color ='black', 
                min_font_size = 10, random_state=100).generate(wc_a) 
  
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off")
plt.xlabel('Word Cloud')
plt.tight_layout(pad = 0) 

print("Word Cloud for Speech (after cleaning)!!")
plt.show()
```
